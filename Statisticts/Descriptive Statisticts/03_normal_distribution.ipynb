```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = np.random.normal(loc=5,scale=2.5,size=1000)
sns.kdeplot(data)
```

```python
mean_data = np.mean(data)
std_data = np.std(data)
mean_data,std_data
```

```python
standardized_data = np.subtract(data, mean_data)/std_data
standardized_mean_data = np.mean(standardized_data)
standardized_std_data = np.std(standardized_data)
standardized_mean_data,standardized_std_data
```

```python
standardized_data = np.ones_like(data)

for i in range(len(data)):
    standardized_data[i] = (data[i] - mean_data)/std_data
```
