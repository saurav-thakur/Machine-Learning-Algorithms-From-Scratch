# Probability Distribution Function

A probability distribution function is a mathematical function that describes
the probability of obtaining different values of a random variable in a particular
probability distribution.

Types:

1. Probability Density Function (PDF)
2. Probability Mass Function (PMF)


Using PDF and PMF we can find the CDF as well which is Cumulative Distribution Function

# Probability Mass Function


PMF stands for Probability Mass Function. It is a mathematical function that
describes the probability distribution of a discrete random variable.
The PMF of a discrete random variable assigns a probability to each possible value
of the random variable. The probabilities assigned by the PMF must satisfy two
conditions:

a. The probability assigned to each value must be non-negative (i.e., greater
than or equal to zero).

b. The sum of the probabilities assigned to all possible values must equal 1.

```python
# Generate a large dataset for PMF calculation
np.random.seed(42)
huge_data = [np.random.randint(0, 10) for _ in range(10000)]  # Simulating 10,000 observations

# Count occurrences of each X value
count_X = Counter(huge_data)

total_observations = sum(count_X.values())

# Compute PMF
pmf = {k: v / total_observations for k, v in count_X.items()}

# Print PMF
for k, prob in sorted(pmf.items()):
    print(f'P(X={k}) = {prob}')
```

```python
# Plot PMF
plt.bar(pmf.keys(), pmf.values(), color='blue', alpha=0.7)
plt.xlabel('X (Number of Occurrences)')
plt.ylabel('P(X)')
plt.title('Probability Mass Function (PMF)')
plt.xticks(range(min(pmf.keys()), max(pmf.keys()) + 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### CDF of PMF

```python
sorted_keys = sorted(pmf.keys())
cdf = {}
cumsum = 0
for i in sorted_keys:
    cumsum += pmf[i]
    cdf[i] = cumsum
cdf
```

```python
# plot cdf

plt.plot(cdf.keys(),cdf.values(),color='red',label="CDF")
plt.xlabel("X (Number of Occurences)")
plt.ylabel("Cumulative Probability")
plt.title("Cumulative Distribution Function (CDF)")
plt.xticks(range(min(cdf.keys()), max(cdf.keys()) + 1))
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.legend()
plt.show()
```

# Probability Density Function


Density estimation is a statistical technique used to estimate the probability
density function (PDF) of a random variable based on a set of observations or data.
In simpler terms, it involves estimating the underlying distribution of a set of data
points.
Density estimation can be used for a variety of purposes, such as hypothesis
testing, data analysis, and data visualization. It is particularly useful in areas such
as machine learning, where it is often used to estimate the probability distribution
of input data or to model the likelihood of certain events or outcomes.

There are various methods for density estimation, including parametric and non-
parametric approaches. Parametric methods assume that the data follows a

specific probability distribution (such as a normal distribution), while non-
parametric methods do not make any assumptions about the distribution and

instead estimate it directly from the data.
Commonly used techniques for density estimation include kernel density
estimation (KDE), histogram estimation, and Gaussian mixture models (GMMs).
The choice of method depends on the specific characteristics of the data and the
intended use of the density estimate.



PDF stands for Probability Density Function. It is a mathematical function that

1. **Why Probability Density and why not Probability?**

In continuous distributions, the probability of a random variable being exactly equal to any particular value is technically 0. This is because there are infinitely many possible values the variable can take, so the probability for any single value is zero.
   - Instead, **probability density** is used to describe the likelihood that the variable falls within a certain range (e.g., between two values). The area under the PDF curve over a given interval gives the probability that the random variable lies within that interval.

   - Since the total probability of the random variable must be 1, the area under the entire PDF curve across all possible values of the random variable equals 1. ## Parametric Density Function

Parametric density estimation involves modeling the underlying distribution of data by assuming it follows a specific known distribution (e.g., Gaussian, Poisson, etc.). The goal is to estimate the parameters of that distribution from the data. Here's how it works:

1. **Assume a distribution**: Choose a probability distribution that you believe best represents your data, such as normal, exponential, or Bernoulli.
   
2. **Estimate parameters**: Using the data, estimate the parameters of the chosen distribution. For instance, for a normal distribution, you'd estimate the mean and variance.

3. **Fit the model**: After estimating the parameters, you can describe the data’s distribution using the fitted distribution. This is the parametric density estimate.

**Common Methods**:
- **Maximum Likelihood Estimation (MLE)**: Find the parameters that maximize the likelihood of observing the given data.
- **Method of Moments**: Match the sample moments (mean, variance, etc.) to the theoretical moments of the chosen distribution.

**Advantages:**
- Can be computationally efficient.
- Provides a well-defined functional form for the probability distribution.

**Disadvantages:**
- The model is restricted by the assumption of a specific distribution, which may not always fit the data well.


```python
np.random.seed(0)

data = np.random.normal(loc=170,scale=10,size=10000)

mean = np.mean(data)
stdev = np.std(data,ddof=1)

x = np.linspace(data.min(),data.max(),len(data))
pdf_fitted = norm.pdf(x,loc=mean,scale=stdev)

plt.hist(data,bins=10,density=True,alpha=0.6,color='green',label="Histogram of Heights")
plt.plot(x,pdf_fitted,'r-',label='Fitted Normal Distribution')
plt.title("Parametric Density Estimation: Normal Distribution")
plt.xlabel("Height")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
from scipy.stats import norm

# Using previously estimated mean (mu_hat) and standard deviation (sigma_hat)
mu_hat = np.mean(data)
sigma_hat = np.std(data, ddof=1)

# Calculate the probability of height between 160 and 170
prob_160_to_170 = norm.cdf(170, loc=mu_hat, scale=sigma_hat) - norm.cdf(160, loc=mu_hat, scale=sigma_hat)

print(f"Probability of height between 160 and 170: {prob_160_to_170:.4f}")
```

# Non Parametric Density Estimation

**Non-parametric density estimation** is a technique used to estimate the probability density function (PDF) of a dataset without making strong assumptions about the underlying distribution. Unlike parametric methods, which assume a specific distribution (e.g., normal distribution), non-parametric methods make fewer assumptions and allow the data to define the shape of the distribution.

The most commonly used non-parametric density estimation method is **Kernel Density Estimation (KDE)**.

### 1. **Kernel Density Estimation (KDE)**



### Steps for Non-Parametric Density Estimation (KDE):
1. **Choose a kernel function** (typically Gaussian).
2. **Choose a bandwidth parameter \( h \)**, which controls the smoothness of the estimated density (a small \( h \) makes the density rough, a large \( h \) makes it too smooth).
3. **Calculate the kernel function for each data point**.
4. **Sum the contributions** from all data points to estimate the density.

### 2. **Advantages of Non-Parametric Density Estimation**:
- **No distribution assumption**: It doesn’t assume any specific form for the underlying distribution, so it's more flexible.
- **Smooth estimation**: KDE provides a smooth estimate of the PDF, which can be useful for visualizing the data distribution.

### 3. **Disadvantages of Non-Parametric Density Estimation**:
- **Bandwidth selection**: The choice of bandwidth \( h \) is crucial and can significantly affect the results. A small bandwidth leads to a jagged estimate, and a large bandwidth leads to an overly smooth estimate.
- **Computationally expensive**: It can be computationally expensive, especially for large datasets, since it requires calculating the kernel for each data point.


```python
data = np.random.normal(size=1000)


x = np.linspace(data.min(),data.max(),num=1000)


def gaussian_kernel(u):
    values = 1/np.sqrt(2*np.pi) * np.exp(-(u**2)/2)
    return values

def kde(data,x,h):

    kde_vals = np.ones_like(data)
    n = len(data)

    for i in range(len(data)):
        u = (x[i] - data)/h
        kde_vals[i] = np.sum(gaussian_kernel(u)/n*h)

    return kde_vals
```

```python
h = 1.2
fitted_data = kde(data,x,h)
```

```python
plt.hist(data,color="green",density=True,label="Histogram of Data")
plt.plot(x,fitted_data,"r-")
```
